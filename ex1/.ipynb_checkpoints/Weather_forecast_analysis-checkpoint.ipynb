## Retrieve Weather data

```python
def get_forecasted_observed(site_id):
    query = "SELECT * FROM weather_data WHERE site_id= {}".format(site_id)
    df = pd.read_sql_query(query,con=engine)
    
    df_day = df[(df.period=='day')]
    df_forecast_day = df_day[df_day.data_type=='forecast']
    df_observed_day = df_day[df_day.data_type=='observation']
    
    #Remove forecasts that are posterior to date of event.
    df_forecast_day = df_forecast_day[df_forecast_day.updated<df_forecast_day.date_hour]
    return df_forecast_day,df_observed_day
```

## Forecasts vs Observed

### Create list of dates ranges

```python
date_from = datetime(2016,11,27)
date_to = datetime(2017,1,7)

def create_datetime_range(date_from,date_to):
    datetime_ranges=[]
    date_of_prediction = date_from
    while date_of_prediction<=date_to:
        date_of_prediction = date_of_prediction
        start_period = date_of_prediction+timedelta(days=1)
        end_period =  date_of_prediction+timedelta(days=7)
        datetime_ranges.append({'date_of_prediction':date_of_prediction,
                               'start_period':start_period,
                               'end_period':end_period})
        date_of_prediction = date_of_prediction+timedelta(7)
        
    return datetime_ranges
```

```python
dates_ranges_list = create_datetime_range(date_from,date_to)
```

### Combine Forecast and Observed values per day

```python
def create_forecasts_vs_observed(dt_ranges,df_forecast_day,df_observed_day):
    ## Forecasts
    df_forecast_day_sunday = df_forecast_day[(df_forecast_day.updated>=dt_ranges['date_of_prediction'])&\
                                         (df_forecast_day.updated<=dt_ranges['date_of_prediction']+timedelta(days=1))]
    #Filter on the latest updated time for each date_hour:
    df_forecast_day_sunday_latest = df_forecast_day_sunday.groupby('date_hour').apply(lambda x:x.ix[x.updated.idxmax()])
    #df_forecast_day_sunday_latest=df_forecast_day_sunday_latest.reset_index()
    
    ## Observations
    df_observed_day_period = df_observed_day[(df_observed_day.date_hour>=dt_ranges['start_period'])&(df_observed_day.date_hour<dt_ranges['end_period'])]
    df_observed_day_period_latest = df_observed_day_period.groupby('date_hour').apply(lambda x:x.ix[x.updated.idxmax()])
    #df_observed_day_period_latest = df_observed_day_period_latest.reset_index()
    
    
    ## Merge forecasts and observations
    df_forecast_vs_observed = pd.merge(df_forecast_day_sunday_latest,df_observed_day_period_latest[['date_hour', u'tt', u'tn',
       u'tx', u'ne', u'ww', u'rrr', u'prrr']],on='date_hour',suffixes=['_forecasted','_observed'])
    df_forecast_vs_observed.drop('data_type',1)
    #df_forecast_vs_observed = df_forecast_vs_observed.set_index('date_hour')
    
    return df_forecast_vs_observed
```

```python
def aggregate_comparison(dates_ranges_list,df_forecast_day,df_observed_day):
    df = pd.DataFrame()
    for dates_ranges in dates_ranges_list:
        try:
            df = pd.concat([df,create_forecasts_vs_observed(dates_ranges,df_forecast_day,df_observed_day)],axis=0)
        except :
            print 'Error on ' ,dates_ranges
    return df
```

```python
def aggregate_comparison_all_sites():
    
    df_all =pd.DataFrame()
    
    for site_dict in ALL_SITES_DICT:
        
        df_forecast_day,df_observed_day = get_forecasted_observed(site_dict['id'])

        df_merged = aggregate_comparison(dates_ranges_list,df_forecast_day,df_observed_day)

        df_merged['site_name']=site_dict['sname']
        df_merged['site_id']=site_dict['id']
        df_merged['idbdsite']=site_dict['idbldsite']

        df_all = pd.concat([df_all,df_merged],axis=0)
        df_all.to_csv('weather_forecast_vs_observed',sep=';',encoding='utf8')

    return df_all
```

```python
df_all = aggregate_comparison_all_sites()
df_all
```


### Compute errors

```python
features_list =['tx','tn','rrr','ne']


def compute_errors(df):
    for feature in features_list:
        df[feature+'_errors']=np.abs(df[feature+'_forecasted']-df[feature+'_observed'])
        

    return df

def compute_accuracy(df):
    for feature in features_list:
        df[feature+'_accuracy'] = np.abs(df[feature+'_errors'])
        df[feature+'_accuracy'] = df[feature+'_accuracy']/df[feature+'_observed']
    return df
```

## Aggregate errors per feature

```python
df_errors = compute_errors(df_all)

def aggregate_all_sites_per_error_value_per_feature(df_errors,feature):
    feature_errors = feature+'_errors'
    df_errors=df_errors[df_errors[feature_errors]<100]
    groupy_errors = df_errors.groupby(['date_hour', feature_errors]).count()
    df_groupy = groupy_errors.unstack(feature_errors).id.fillna(0)
    return df_groupy

```

```python
start = df_groupy.index[0]
end = df_groupy[0].index[-1]

print start,end
```

```python
for feature in features_list:
    feature_errors = feature+'_errors' 
    df_errors_feature = df_errors[df_errors[feature_errors]<100][['date_hour',feature_errors]]
    df_error_feature_groupy = df_errors_feature.groupby('date_hour').mean()
    df_error_feature_groupy.plot(title='Average absolute errors  for all sites')
    for date_dict  in dates_ranges_list:

        plt.axvline(date_dict['date_of_prediction'],color='#a30037',ls='--')
    
```

```python
groupy_tx_errors = aggregate_all_sites_per_error_value_per_feature(df_errors,'tx')
groupy_tn_errors= aggregate_all_sites_per_error_value_per_feature(df_errors,'tn')
groupy_ne_errors= aggregate_all_sites_per_error_value_per_feature(df_errors,'ne')
groupy_rrr_errors= aggregate_all_sites_per_error_value_per_feature(df_errors,'rrr')
```

```python
def plot_errors_aggregated(df_groupy,title_chart,max_value):
    fig, ax = plt.subplots()
    colors = cm.Blues_r(np.linspace(0, 1, (df_groupy.shape[1])))
    for i in df_groupy.columns:
        ax.scatter(df_groupy.index,np.full((1, df_groupy.shape[0]), i),s=df_groupy[i]/2, alpha=1,color=colors[0])
    ax.set_xlim([start, end])
    title_chart = 'Errors for '+title_chart
    ax.set_title(title_chart)
    for date_dict  in dates_ranges_list:

        ax.axvline(date_dict['date_of_prediction'],color='#a30037',ls='--')
    ax.set_ylim([-1, max_value])
   
```

```python
plot_errors_aggregated(groupy_tx_errors,'temperature max',10)
```

```python
plot_errors_aggregated(groupy_tn_errors,'temperature min',10)
```

```python
plot_errors_aggregated(groupy_ne_errors,'precipitation',10)
```

```python
plot_errors_aggregated(groupy_rrr_errors,'weather situtation',10)
```
